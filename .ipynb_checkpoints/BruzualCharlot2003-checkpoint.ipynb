### Extract synthetic stellar population spectra
This step utilizes a Python wrapper [originally developed](https://github.com/vihangmehta/extract_BC03) originally developed by Vihang Mehta (and [modified](https://github.com/hughdickinson/extract_BC03) by Hugh Dickinson). It invokes original [GALAXEV](http://www.bruzual.org/bc03/) package of FORTRAN utilities developed to support the synthetic stellar population atlases described in [Bruzual & Charlot 2003](http://mnras.oxfordjournals.org/content/344/4/1000.full.pdf).

```python
template = TemplateSED_BC03(metallicity=0.02, age=[1,2,3,5,10], sfh='single', tau=0.5, Av=1,
                            dust='none', emlines=False,
                            redshift=2, igm=True,
                            imf='chab', res='hr', uid='test', units='flambda',
                            rootdir=srcRootDir, modelsdir=modelRootDir,workdir=workDirRoot,
                            library_version=2016, library='xmiless',
                            cleanup=True,verbose=True)
```

```python
template.generate_sed()
```

#### Store the extracted spectra in a pandas DataFrame to simplify manipulation

```python
sedFrame = pd.DataFrame(template.sed)
luminosityDensityColNames = [col for col in sedFrame.columns.values if 'spec' in col]
```

#### Scale luminosity densities from their output ($1{\rm M_{\odot}}$-equivalent) normalization to $10^{11}{\rm M_{\odot}}$-equivalent

```python
logMassSolarUnits = 10
for col in luminosityDensityColNames :
    sedFrame[col] = sedFrame[col]*(10**logMassSolarUnits)/template.M_unnorm[col]
```

#### Explicitly define the units of the wavelength as Ã…ngstroms

```python
sedFrame['rest_wavelength'] = (sedFrame['rest_wavelength'].values*astrounits.angstrom)
```

## Manipulate the synthetic spectra to compute the apparent and absolute magnitudes
Use values and functions `astropy.cosmology` to compute cosmological parameters and scalings for the Planck 2015 best-fit cosmology.

```python
print(astropy.cosmology.Planck15.__doc__)
```

### Compute *observed* wavelength values that correspond to each *luminosity* density

```python
gc.collect()

sedFrame['observed_wavelength'] = sedFrame['rest_wavelength']/(1.0+template.redshift)
axes = sedFrame.plot.line(x='observed_wavelength', y=luminosityDensityColNames, logy=True, logx=True)
axes.set_xlabel(r'Observed Wavelength [${\rm\AA}$]')
axes.set_ylabel(r'Luminosity density [${\rm erg\;s^{-1}\;\AA^{-1}}$]')
```

### Compute the redshift-corrected flux *density*
For a luminosity spectrum expressed in erg s$^{-1}$ Ã…$^{-1}$ this computation is defined as:
$$f_{\lambda} = \frac{L_{\lambda}[\lambda\cdot(1+z)^{-1}]}{(1+z)\cdot4\pi d_{L}^{2}}$$

where $\lambda$ is the *observed* wavelength and $d_{L}$ is the *luminosity distance*.

#### Compute luminosity distance in cm.

```python
## retrieve in cm to make units correct for flambda -> fnu conversion below
luminosityDistance = Planck15.luminosity_distance(template.redshift).to(astrounits.cm)
print('luminosityDistance = {}'.format(luminosityDistance))
```

#### Compute the *constant* denominator of the flux density expression.

```python
correctionDenominator = (1.0+template.redshift)*4.0*np.pi*luminosityDistance**2
print('correctionDenominator = {}'.format(correctionDenominator))
```

#### Compute the *flux* densities.

```python
fluxDensityColNames = []
for specName in luminosityDensityColNames :
    fluxDensityColNames.append(specName + '_flux_density')
    sedFrame[fluxDensityColNames[-1]] = sedFrame[specName]/correctionDenominator
axes = sedFrame.plot.line(x='observed_wavelength', y=fluxDensityColNames, logy=True, logx=True)
axes.set_xlabel(r'Observed Wavelength [${\rm\AA}$]')
axes.set_ylabel(r'Flux density [${\rm erg\;s^{-1}\;cm^{-2}\;\AA^{-1}}$]')
axes.legend(loc='best', prop={'size':10})
```

### Map the flux densities to frequency space i.e. compute $f_{\lambda}\mapsto f_{\nu}$
For $f_{\lambda}$ defined in ${\rm erg\;s^{-1}\;cm^{-2}\;\unicode{xC5}^{-1}}$, $f_{\nu}$ in Jansky units is computed using:
$$\frac{f_{\nu}}{\rm Jy} = 3.34\times10^{4}\left(\frac{\lambda_{\rm pivot}}{\unicode{xC5}}\right)^{2}\frac{f_{\lambda}}{\rm erg\;s^{-1}\;cm^{-2}\;\unicode{xC5}^{-1}}$$

Where $\lambda_{\rm pivot}$ is the *pivot wavelength* for the filter bandpass and can be retrieved from e.g.the online [Filter Profile Service](http://svo2.cab.inta-csic.es/svo/theory/fps3/index.php?mode=browse&gname=HST&gname2=WFC3_IR)

```python
filterPivotWavelengths = {110 : 11534.46*astrounits.angstrom,
                          160 : 15370.33*astrounits.angstrom}
```

```python
filterNumber = 160
```

```python
fluxDensityJyColNames = []
for specName in fluxDensityColNames :
    fluxDensityJyColNames.append(specName + '_flux_density_jansky')
    pivotLambda = filterPivotWavelengths[filterNumber]
    sedFrame[fluxDensityJyColNames[-1]] = 3.34e4*(pivotLambda*pivotLambda).value*sedFrame[specName]
axes = sedFrame.plot.line(x='observed_wavelength', y=fluxDensityJyColNames, logy=True, logx=True)
axes.set_xlabel(r'Observed Wavelength [${\rm\AA}$]')
axes.set_ylabel(r'Flux density [${\rm Jy}$]')
axes.legend(loc='best', prop={'size':10})
```

#### Convert observed wavelengths to frequencies

```python
sedFrame['observed_wavelength_m'] = sedFrame['observed_wavelength'].apply(lambda wavelength : (wavelength*astrounits.angstrom).to(astrounits.m).value)
sedFrame['observed_frequency'] = sciconsts.c/sedFrame['observed_wavelength_m']
axes = sedFrame.sort_values(by='observed_frequency').plot.line(x='observed_frequency', y=fluxDensityJyColNames, logy=True, logx=True)
axes.set_xlabel(r'Observed Frequency [${\rm Hz}$]')
axes.set_ylabel(r'Flux density [${\rm Jy}$]')
axes.legend(loc='best', prop={'size':10})
```

## Fold the spectra with the instrument filter response

```python
responseFiles = {110 : '/Users/hughdickinson/Documents/Physics/WISPS_Hubble_Constraint/HST_Responses/HST_WFC3_IR.F110W.dat.txt',
                 160 : '/Users/hughdickinson/Documents/Physics/WISPS_Hubble_Constraint/HST_Responses/HST_WFC3_IR.F160W.dat.txt'}

responseWavelengths = []
responseValues = []
response = pd.read_csv(responseFiles[filterNumber], header=None, names=['wavelength', 'response'], delim_whitespace=True)
```

#### Explicitly define the wavelength units to be Ã…ngstrom

```python
response['wavelength'] = (response['wavelength'].values*astrounits.angstrom)
```

#### Convert response to frequency units

```python
response['wavelength_m'] = response['wavelength'].apply(lambda wavelength : (wavelength*astrounits.angstrom).to(astrounits.m).value)
response['frequency'] = sciconsts.c/response['wavelength_m']
```

```python
axes = response.sort_values(by='frequency').plot.line(x='frequency', y='response', logx=True)
axes.set_xlabel(r'Observed Frequency [${\rm Hz}$]')
axes.set_ylabel(r'Response function')
axes.legend(loc='best', prop={'size':10})
```

#### Instantiate interpolator to compute a response value at arbitrary frequency.

```python
interpResponse = sp.interpolate.InterpolatedUnivariateSpline(response.sort_values(by='frequency').frequency.values, 
                                                             response.sort_values(by='frequency').response.values, 
                                                             ext='zeros')
```

```python
fluxDensityJyWithResponseColNames = []
for specName in fluxDensityJyColNames :
    fluxDensityJyWithResponseColNames.append(specName + '_withResponse')
    sedFrame[fluxDensityJyWithResponseColNames[-1]] = sedFrame[specName] * interpResponse(sedFrame['observed_frequency'])
```

```python
axes = sedFrame.sort_values(by='observed_frequency').plot.line(x='observed_frequency', y=fluxDensityJyWithResponseColNames, logy=True, logx=True, xlim=(np.min(response['frequency'].values), np.max(response['frequency'].values)))
axes.legend(loc='best', prop={'size':9})
```

### Compute magnitude
This requires integration over the response-folded *flux* distribution. The reuired integrand values are obtained by dividing each by their corresponding frequency value.

```python
fluxDensityJyWithResponseIntegrandColNames = []
for specName in fluxDensityJyWithResponseColNames :
    fluxDensityJyWithResponseIntegrandColNames.append(specName + '_integrand')
    sedFrame[fluxDensityJyWithResponseIntegrandColNames[-1]] = sedFrame[specName] / sedFrame['observed_frequency']
axes = sedFrame.sort_values(by='observed_frequency').plot.line(x='observed_frequency', y=fluxDensityJyWithResponseIntegrandColNames, logy=True, logx=True, xlim=(np.min(response['frequency'].values), np.max(response['frequency'].values)))
axes.legend(loc='best', prop={'size':8})
```

```python
response['response_integrand'] = response['response']/response['frequency']
axes = response.sort_values(by='frequency').plot.line(x='frequency', y='response_integrand', logx=True)
axes.set_xlabel(r'Observed Frequency [${\rm Hz}$]')
axes.set_ylabel(r'Response function integrand')
axes.legend(loc='best', prop={'size':10})
```

#### Integrate the response integrand and response-weighted flux integrands

```python
responseIntegral = sp.integrate.simps(x=response.frequency.values, y=response.response_integrand.values)
specIntegrals = []
for specName in fluxDensityJyWithResponseIntegrandColNames :
    specIntegrals.append(sp.integrate.simps(x=sedFrame['observed_frequency'].values, y=sedFrame[specName].values))
```

#### Compute magnitude from flux integrals 
Using the adopted units, the apparent AB magnitude is computed using:
$$m_{\rm AB} = -2.5\log_{10}\left(\frac{F}{R}\right) + 8.9$$

where $F$ and $R$ are the integrals of the response-weighted flux distributions and the response, respectively. The factor of 8.9 derives from integration of the AB magnitude's defined "equal-energy" filter response function using the defined zero point flux density of 3631 Jy. 

```python
luminosityDistancePc = luminosityDistance.to(astrounits.pc)
appABMagnitudes = []
absABMagnitudes = []
for integral in specIntegrals :
    appABMagnitude = -2.5*np.log10(integral/responseIntegral) + 8.9 
    distanceModulus = 5.0*np.log10(luminosityDistancePc.value) - 5.0
    absABMagnitude = appABMagnitude - distanceModulus
    appABMagnitudes.append(appABMagnitude)
    absABMagnitudes.append(absABMagnitude)

magnitudeDict = {'ageGyr' : template.age, 'appABMagnitude' : appABMagnitudes, 'absABMagnitude' : absABMagnitudes}
magnitudeFrame = pd.DataFrame.from_dict(magnitudeDict)
axes = magnitudeFrame.plot.line(x='ageGyr', y='appABMagnitude')
axes.set_xlabel(r'Starburst age at z={} [Gyr]'.format(template.redshift))
axes.set_ylabel(r'$m_{\rm AB}$', size=15)
axes.legend(loc='best', prop={'size':10})
axes.set_title('Assumed stellar mass $M_{}=10^{}\;M_{}$'.format('{\star}','{' + str(logMassSolarUnits) + '}','{\odot}'))
```
